```python
df['brand'].value_counts().plot(kind='bar',
                              title='count by brand',
                              xlabel='brand',
                              ylabel='count',
                              color='violet')
                        
```

```python
df['processor_brand'].value_counts().plot(kind='bar',
                              title='count by processor_brand',
                              xlabel='processor_brand',
                              ylabel='count',
                              color='violet')
```

## 2.continuous analysis

```python
for i in con:
    sns.histplot(data=df,x=i,kde=True,color= 'green')
    plt.title(f'Histogram for {i}')
    plt.show()
```

```python
x=1
plt.figure(figsize=(12,12))
for i in con:
    plt.subplot(2,2,x)
    x=x+1
    sns.histplot(data=df,x=i,kde=True)
    plt.title(f'Histogram for {i}')
plt.show()
```
